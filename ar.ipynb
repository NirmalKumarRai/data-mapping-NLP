```python
## exact match
import pandas as pd
from nltk.stem import PorterStemmer

# Sample file paths, replace with your actual file paths
file_path1 = r'D:\raptor supplies\22nov\na item name.csv'
file_path2 = r'D:\raptor supplies\22nov\Duplicate Item cluster keyword (1).csv'

# Read the CSV files into DataFrames
df1 = pd.read_csv(file_path1)
df2 = pd.read_csv(file_path2)

# Extract relevant columns
BC = df1['raw_value']
item = df2['Keyword']

# Initialize stemmer
stemmer = PorterStemmer()

# Create a function to check if stemmed words in element2 are exact matches in element1
def is_exact_match(element1, element2):
    words1 = set([word.lower() for word in element1.split()])
    words2 = set([word.lower() for word in element2.split()])
    return words1 == words2

# Open the file in write mode to create or overwrite it
with open("exact_item_map.csv", "a") as f:
    for item1 in BC:
        matches = [item2 for item2 in item if is_exact_match(item1, item2)]
        result_str = f"{item1}, {' | '.join(matches)}\n"
        f.write(result_str)
        print(result_str)
```
